```python
import numpy as np
from numpy.linalg import norm

A = np.array([2,1,2,3,2,9])
B = np.array([3,4,2,4,5,5])

#compute cosine similarity
cosine = np.dot(A,B)/(norm(A)*norm(B))
print("cosine Similarity:",cosine)
```

```python
from transformers import pipeline

summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

text ="""The  global shift toward renewable energy is accelerating,
driven by concerns over climate change and energy security. Solar and
wind power installations have reached record levels worldwide."""

summary = summarizer(text, max_length=30, min_length=10, do_sample=False)
print(summary[0])
```

```python
from transformers import AutoTokenizer
# load a tokenizer (e.g,BERRT )
tokeninzer = AutoTokenizer.from_pretrained("bert-base-uncased")

text ="Machine Learning is transforming materials science!"
# Tokenize
tokens = tokeninzer.tokenize(text)
token_ids = tokeninzer.encode(text)

print("tokens:",tokens)
print("token IDs:",token_ids)
```
